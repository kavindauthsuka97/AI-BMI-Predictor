```python
import boto3
from botocore.exceptions import NoCredentialsError, ClientError

# --- Config ---
bucket_name = "ai-bmi-predictor"
local_file_path = "vgg16_front_masked_features.pkl"  # adjust if it's in another directory
s3_key = "feature-extraction-data/training-data/vgg16_front_masked_features.pkl"

# --- Create S3 client (SageMaker usually has an IAM role attached so no keys needed) ---
s3 = boto3.client("s3")

try:
    # This will automatically create the "folders" (prefixes) if they don't exist
    s3.upload_file(local_file_path, bucket_name, s3_key)
    print(f"Uploaded {local_file_path} to s3://{bucket_name}/{s3_key}")
except FileNotFoundError:
    print(f"Local file not found: {local_file_path}")
except NoCredentialsError:
    print("AWS credentials not available. Check your IAM role / configuration.")
except ClientError as e:
    print(f"Unexpected error: {e}")
```
