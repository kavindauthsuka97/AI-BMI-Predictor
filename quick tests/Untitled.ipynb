* load training dataset

```python
import boto3
import pandas as pd

# --- S3 information ---
bucket = "amazon-bodym"                 # your S3 bucket name
key = "train/subject_to_photo_map.csv"  # key inside the bucket

# --- Read CSV directly from S3 ---
s3 = boto3.client("s3")

obj = s3.get_object(Bucket=bucket, Key=key)
data_train = pd.read_csv(obj["Body"])

data_train.head()
```

* load testA dataset

```python
import boto3
import pandas as pd

# --- S3 information ---
bucket = "amazon-bodym"                 # your S3 bucket name
key = "testA/subject_to_photo_map.csv"  # key inside the bucket

# --- Read CSV directly from S3 ---
s3 = boto3.client("s3")

obj = s3.get_object(Bucket=bucket, Key=key)
data_testA = pd.read_csv(obj["Body"])

data_testA.head()
```

* load testB dataset

```python
import boto3
import pandas as pd

# --- S3 information ---
bucket = "amazon-bodym"                 # your S3 bucket name
key = "testB/subject_to_photo_map.csv"  # key inside the bucket

# --- Read CSV directly from S3 ---
s3 = boto3.client("s3")

obj = s3.get_object(Bucket=bucket, Key=key)
data_testB = pd.read_csv(obj["Body"])

data_testB.head()
```

```python
# Ensure subject_id columns exist
assert 'subject_id' in data_train.columns, "data_train has no subject_id column"
assert 'subject_id' in data_testA.columns, "data_testA has no subject_id column"

# convert to sets for fast lookup
train_ids = set(data_train['subject_id'])
test_ids  = set(data_testA['subject_id'])

# subjects that ARE in train
test_in_train = test_ids.intersection(train_ids)

# subjects that are NOT in train
test_not_in_train = test_ids.difference(train_ids)

print("Number of test subjects:", len(test_ids))
print("Subjects found in train:", len(test_in_train))
print("Subjects NOT found in train:", len(test_not_in_train))
```
