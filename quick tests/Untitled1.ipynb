```python
import boto3

s3 = boto3.resource('s3')
bucket_name = "ai-bmi-predictor"
bucket = s3.Bucket(bucket_name)

# Prefixes you want to KEEP
keep_prefixes = [
    "data/",
    "feature-extraction-data/",
    "image-segmentation/",
    "test-data/",
    "trained-models/"
]

# Function to check whether an object key starts with any keep prefix
def should_keep(key):
    return any(key.startswith(prefix) for prefix in keep_prefixes)

objects_to_delete = []

# Identify all objects that should NOT be kept
for obj in bucket.objects.all():
    if not should_keep(obj.key):
        objects_to_delete.append({'Key': obj.key})

# Delete in batches (S3 API limit = 1000 per call)
if objects_to_delete:
    print(f"Deleting {len(objects_to_delete)} objects...")
    for i in range(0, len(objects_to_delete), 1000):
        batch = objects_to_delete[i:i+1000]
        bucket.delete_objects(Delete={'Objects': batch})
    print("Deletion complete.")
else:
    print("No objects to delete. All objects match keep prefixes.")
```
