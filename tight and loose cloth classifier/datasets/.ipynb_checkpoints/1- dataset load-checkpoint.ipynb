```python
# Install gdown if not already installed
!pip install --quiet gdown

import gdown

# Google Drive file URL
url = "https://drive.google.com/uc?id=1hWR99dYtFgBIFN8ZB1fx1h7dammeuEqw"

# Output file name (zip will be saved in current working directory)
output = "dataset.zip"

# Download
gdown.download(url, output, quiet=False)
```

* unzip the zip file

```python
import zipfile

with zipfile.ZipFile("dataset.zip", 'r') as zip_ref:
    zip_ref.extractall("./dataset")
```

```python
import os
import boto3

s3 = boto3.client("s3")

bucket_name = "ai-bmi-predictor"
s3_base_path = "tight and loose classifier/orginal dataset"

local_base_path = "./dataset"

for root, dirs, files in os.walk(local_base_path):
    for file in files:
        local_file_path = os.path.join(root, file)

        # Preserve directory structure inside S3
        relative_path = os.path.relpath(local_file_path, local_base_path)
        s3_key = f"{s3_base_path}/{relative_path}"

        s3.upload_file(local_file_path, bucket_name, s3_key)
        print(f"Uploaded: s3://{bucket_name}/{s3_key}")
```
